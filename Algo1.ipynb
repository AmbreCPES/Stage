```julia
function M1!(du, u, p, t)
    z = u[1] + u[2] + u[3]
    du[1] = p[2] * u[1] * (p[1] - z) - p[3] * u[1]
    du[2] = p[3] * u[1] + p[4] * u[2] * (p[1] - z) - p[5] * u[2]
    du[3] = p[4] * u[2]  - p[6] * u[3]
end
```

```julia
using DifferentialEquations
u0 = [1.0; 0.0; 0.0]
tspan = (0.0, 1.0)
p = [1, 0.5, 0.5, 0.5, 0.5, 0.5 ]
prob = ODEProblem(M1!, u0, tspan, p)
sol = solve(prob)
```

```julia
plot(sol)
```
